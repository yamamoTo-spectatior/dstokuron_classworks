```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import japanize_matplotlib
```

```python
sisan = 100
term = 10*12
sarary = 10
sarary1 = 50
rate = 0.05
rate1 = 0.01
```

```python
s_sisans = [sisan]
s_sisans1 = [sisan]
r_sisans = [sisan]
r_sisans1 = [sisan]
s_sisan = sisan
s_sisan1 = sisan
r_sisan = sisan
r_sisan1 = sisan
for i in range(term-1):
    s_sisan += sarary
    s_sisan1 += sarary1
    r_sisan += r_sisan*rate
    r_sisan1 += r_sisan1*rate1
    s_sisans.append(s_sisan)
    s_sisans1.append(s_sisan1)
    r_sisans.append(r_sisan)
    r_sisans1.append(r_sisan1)
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.5)
#fig.suptitle("all graph")

ax1 = fig.add_subplot(111, title="",xlabel="()", ylabel="()", yscale=None)
# axesplot
ax1.plot(range(term), s_sisans, "--", c="#000000", linewidth=1, marker=None)
ax1.plot(range(term), s_sisans1, "--", c="#000000", linewidth=1, marker=None)
ax1.plot(range(term), r_sisans, "-", c="#000000", linewidth=2, marker=None)
ax1.plot(range(term), r_sisans1, "-", c="#000000", linewidth=2, marker=None)
ax1.text(0,s_sisans[0]+1000,f"{s_sisans[0]}")
ax1.text(term,r_sisans[-1]+1000,f"{int(r_sisans[-1])}")
ax1.text(term,s_sisans1[-1]+1000,f"{s_sisans1[-1]}")
ax1.text(term,s_sisans[-1]+100,f"{s_sisans[-1]}")
ax1.text(term,r_sisans1[-1]+100,f"{int(r_sisans1[-1])}")

ax1.legend(['(10)','(50)', '(5%)', '(1%)'])  # 

fig.show()
```
